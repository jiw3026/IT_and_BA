```python
#파일 불러오기
from google.colab import files
myfile = files.upload()
```

```python
import io
import pandas as pd
```

```python
data = pd.read_csv(io.BytesIO(myfile['health_data.csv']))
data.head()
```

```python
#시력(독립) - 흡연(종속)
#위 주제에 관해, 세 개의 분류모델을 구성할 예정
```

```python
#흡연상태 - 시력 간 관계 파악을 위한 분석 진행
#(1) 이를 위해 좌우 시력의 평균을 계산한 열 추가
data['total'] = data['시력(좌)'] + data['시력(우)']
data['시력평균'] = data['total'] / 2 
data.head()
```

```python
data.head()   #기준년도, 가입자일련번호, 이름 바꾼 average 등 3개 열 삭제
```

```python
#시력 칼럼은 1미만, 1 이상으로 변경
data['eye'] = 0
```

```python
data.loc[data['시력평균']<1,'eye']=0
data.loc[data['시력평균']>=1,'eye']=1
data.head()  #eye에 해당 결과가 적용되었음을 확인할 수 있다. 
```

```python
#기존 시력(좌),(우) 셀은 삭제
data.head()
```

```python
data2 = pd.get_dummies(data)
data2.head()
```

```python
data2.head()
```

```python
data2.loc[data2['흡연상태_No']==1,'smoke']=0
data2.loc[data2['흡연상태_NotNow']==1,'smoke']=0
data2.loc[data2['흡연상태_Yes']==1,'smoke']=1
data2.tail() #smoke에 해당 결과가 적용되었음을 확인할 수 있다. 
```

```python
data2 = data2.drop(['기준년도', '가입자일련번호','데이터공개일자_12/29/2021'],axis=1)
```

```python
data2 = data2.dropna() #결측치가 있는 경우 분류모델 형성이 안 되나, 하나하나 따져가며 제거하기엔 데이터셋이 너무 크기에 결측치 있는 행 모두 제거. 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(data2.drop('smoke', 1), data2['smoke']) #test, train 세트로 분류
```
